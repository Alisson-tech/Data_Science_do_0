## Gráfico de histyrogramas agrupados - pg 48

```python
from matplotlib import pyplot as plt
from collections import Counter
```

```python
grades = [12, 95, 25, 60, 68, 74, 39, 82, 100, 54, 34]

#agrupar notas por decimais e colocar o 100 com 90
histogram = Counter(min(grade // 10 * 10, 90) for grade in grades)
print(histogram)
```

```python
plt.bar([x + 5 for x in histogram.keys()], #mova as barras para a direita em 5
       histogram.values(),                 # altura da barra (corresponde a quantidade de valor igual)
       10,                                 # 10 de largura por barra
       edgecolor=(0,0,0))                  #cor da borda da barra

plt.axis([-5, 105, 0, 5]) # intervalo de largura e altura
plt.xticks([10 * i for i in range(11)]) # eixo x de 0 a 100 em 10 em 10 
plt.xlabel('Notas estudantes')
plt.ylabel('nº estudantes')
plt.title('Distribuição de notas avaliação 1')
plt.show()
```

## Gráfico de linhas - pg 50 

```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
square = [256, 128, 64, 32, 16, 8, 4, 2, 1 ]

total_error = [x + y for x, y in zip(variance, square)] #Soma cada item da variance com square
total_error
```

```python
xs = [i for i, _, in enumerate(variance)]
xs
```

```python
plt.plot(xs, variance, 'g-', label='variance') #gráfico com linha verde solida
plt.plot(xs, square, 'r-.', label='square^2') # linha vermelha de ponto tracejado
plt.plot(xs, total_error, 'b:', label='erro total') #linha pontilhada azul

#numero corresponde a localização da legenda
plt.legend(loc=9)

plt.xlabel('Complexidade do modelo')
plt.xticks([])
plt.title('O dilema Viés-Variance')
plt.show()
```

## Gráfico de dispersão - pg 51

```python
qtd_friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(qtd_friends, minutes)

#rotular cada ponto

for label, friend_count, minutes_count in zip(labels , qtd_friends, minutes):
    plt.annotate(label, xy=(friend_count, minutes_count), #Coloque o rotulo no ponto
                xytext=(5, -5),                           # Levemente deslocado 
                 textcoords='offset points' 
                )
plt.title('Minutos Diários x Número de amigos')
plt.xlabel('nº amigos')
plt.ylabel('Minutos no Site')
plt.show()
```
